# Song Recommender - Web Scrapping

PART I

The goal of this lab is to create a function: scrape_hot100() to scrape the current top 100 songs present at https://www.billboard.com/charts/hot-100 and their respective artists, put the information into a pandas dataframe, and save the dataframe in a csv file in the current folder.

```python
# Importing libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Setting up the url in a variable
url = "https://www.billboard.com/charts/hot-100/"
```

```python
# Downloading html with a get request
response = requests.get(url)
response.status_code
```

```python
# Parsing HTML
soup = BeautifulSoup(response.content, "html.parser")
#soup.prettify
```

```python
# Retrieving the desired info
soup.select('#post-1479786 > div.pmc-paywall > div > div > div > div.chart-results-list.\/\/.lrv-u-padding-t-150.lrv-u-padding-t-050\@mobile-max')
```

```python
# What is the path to song titles ?
soup.select("h3.c-title.a-no-trucate")
```

```python
soup.select("h3.c-title.a-no-trucate")[0].get_text()
```

```python
# creating the list of titles
titles = []
num_iter = len(soup.select("h3.c-title.a-no-trucate"))
num_iter
```

```python
for i in range(num_iter):
    titles.append(soup.select("h3.c-title.a-no-trucate")[i].get_text())
print(titles)
```

```python
# Finding the path to artists :
soup.select("span.c-label.a-no-trucate") 
```

```python
soup.select("span.c-label.a-no-trucate")[0].get_text()
```

```python
# creating the list of titles
artists = []
num_iter2 = len(soup.select("span.c-label.a-no-trucate"))
num_iter2
```

```python
for i in range(num_iter2):
    artists.append(soup.select("span.c-label.a-no-trucate")[i].get_text())
print(artists)
```

```python
#Creating the dataframe with song titles and artists :
top_100 = pd.DataFrame({"title":titles,
                       "artist":artists
                      })

top_100.head()
```

```python
type(top_100)
```

```python
top_100.tail()
```

```python
top_100.columns
```

```python
top_100['title'] = top_100['title'].apply(lambda x: x.strip('\n\t')) 
```

```python
top_100['artist'] = top_100['artist'].apply(lambda x: x.strip('\n\t')) 
```
